```python
# Read the 'n.a','Nat available' as Nan value from csv
# This part will convert negative revenue to NAN using dictionary
import pandas as pd
df = pd.read_csv('stock_data.csv',na_values={
    'eps' : ['n.a.','not available'],
    'revenue' : ['n.a.','not available',-1],
    'people' : ['n.a.','not available'],
    'price' : ['n.a.','not available']
})
df
```
